```python
from pathlib import Path
import csv
import brixs as br
# create energy & intensity arrays
energy = []
intensity = []
# write complete filepath
filepath = Path(r"complete filepath")
#open .csv 
with open(filepath) as csvfile:
     reader = csv.reader(csvfile, delimiter=',')
     for i, row in enumerate(reader):
          # eliminates the first line
          if i >= 1:
              # converts string to float
              x_values = float(row[0])
              y_values = float(row[1])
              # write columns for Spectrum style
              energy.append(x_values)
              intensity.append(y_values)
# create Spectrum object based on columns
s = br.Spectrum(x =energy,y=intensity)
# normalize and apply calib in spectrum
s_max = max(s.y)
s = s.set_factor(1/s_max).set_calib(-1)
```
